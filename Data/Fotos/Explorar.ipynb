```python
import os
import pandas
```

```python
files=os.listdir()
explorar=None
for i in files:
    if 'Tipo'in i:
        pre_file=pandas.DataFrame(os.listdir(i),columns=['file'])
        pre_file['Tipo']=str(i).replace('Tipo ','')
        explorar=pandas.concat([explorar,pre_file])
explorar=pandas.get_dummies(explorar,columns=['Tipo'],dtype=int)
explorar
```

```python
explorar.to_excel('Explorar.xlsx',index=False,sheet_name='explorar')
```
